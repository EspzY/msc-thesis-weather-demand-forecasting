# 10 LightGBM Full-Step5 Results Audit and Thesis Outputs

This notebook is the reporting and reproducibility layer for the completed notebook 09 `full_step5` LightGBM rolling-origin evaluation. It reads existing `full_step5` artifacts and rebuilds thesis-ready tables, figures, audit files and documentation outputs.

The notebook does not train models, tune hyperparameters, run Optuna, run SHAP, or compare against smoke, benchmark or alpha-sensitivity runs.


## Inputs and output contract

The required inputs are the `full_step5` row-level prediction parquet and saved metrics CSV under `outputs/lightgbm_rolling_origin/`. Other `full_step5` outputs are used when available; missing optional files are reported and recomputed from predictions where possible.

Outputs are written to the `full_step5_audit` folder, thesis table folders, and the corresponding documentation files for the notebook 09 full-step evaluation.


```python
import json
import runpy
from pathlib import Path

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "README.md").exists():
    # Resolve the project root without hardcoded local paths.
    PROJECT_ROOT = Path.cwd().parent

assert (PROJECT_ROOT / "README.md").exists(), "Could not locate project root."
SCRIPT_PATH = PROJECT_ROOT / "tools" / "build_full_step5_results_audit.py"
assert SCRIPT_PATH.exists(), SCRIPT_PATH

print("Project root:", PROJECT_ROOT)
print("Audit helper:", SCRIPT_PATH)
```

## Preflight discovery

This cell lists `full_step5` artifacts in `outputs/lightgbm_rolling_origin/`. It excludes smoke outputs, notebook 07 benchmark outputs and notebook 09b alpha-sensitivity outputs.


```python
base = PROJECT_ROOT / "outputs" / "lightgbm_rolling_origin"
full_step5_files = sorted(p.name for p in base.glob("*full_step5*") if p.is_file())
assert any("predictions_full_step5.parquet" in name for name in full_step5_files), (
    "Missing full_step5 predictions parquet."
)
assert any("metrics_full_step5.csv" in name for name in full_step5_files), (
    "Missing full_step5 metrics CSV."
)

print("full_step5 files discovered:", len(full_step5_files))
for name in full_step5_files:
    print("-", name)
```

## Build audit outputs

The helper script recomputes metrics from saved predictions, validates saved metrics, builds thesis-ready tables and figures, writes audit reports and thesis snippets, and updates the Markdown documentation layer. It is intended to be safe to rerun after notebook 09 regenerates the `full_step5` outputs.


```python
runpy.run_path(str(SCRIPT_PATH), run_name="__main__")
```

## Documentation updates

The helper script updates project documentation sections related to notebook 09 `full_step5`, notebook 10 outputs and the audit changelog.


```python
audit_dir = PROJECT_ROOT / "outputs" / "lightgbm_rolling_origin" / "full_step5_audit"
summary_path = audit_dir / "full_step5_audit_summary.json"
docs_read_path = audit_dir / "tables" / "full_step5_docs_read_status.csv"
docs_updated_path = audit_dir / "tables" / "full_step5_docs_updated.csv"

print("Summary:")
print(json.dumps(json.loads(summary_path.read_text(encoding="utf-8")), indent=2))

print("\nDocumentation files read/status table:", docs_read_path)
print("Documentation files updated table:", docs_updated_path)
print("Changelog:", PROJECT_ROOT / "docs" / "FULL_STEP5_AUDIT_CHANGELOG.md")
```
